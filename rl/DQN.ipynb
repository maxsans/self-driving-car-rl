```python
from gym.game_env import CarRacingEnv

from stable_baselines3 import DQN
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.monitor import Monitor
```

#### Create the environment

_A vectorized environment is used to allow multiple parallel environments to be run_

```python
env = CarRacingEnv(render_mode="human")
```

#### Create the model

```python
model = DQN(
    "MlpPolicy",
    env,
    batch_size=64,
    verbose=1
)
```

#### Train the model

```python
model.learn(total_timesteps=100_000)
```

#### Evaluate the model

```python
mean_reward, std_reward = evaluate_policy(model, model.get_env(), n_eval_episodes=1)
print(f"mean_reward:{mean_reward:.2f} +/- {std_reward:.2f}")
```

#### Make the model run in a human viewable window

```python
env = CarRacingEnv(render_mode="human")

obs, info = env.reset()
while True:
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = env.step(action)
    env.render()
    if terminated or truncated:
        obs, info = env.reset()
    
```


#### Save the model

```python
model.save("dqn_car_racing")
```
